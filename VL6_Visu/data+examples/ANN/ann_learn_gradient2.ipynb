# Gradient Descent - simpel und schnell verstanden

```python
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
def func_id(x):
    return x
```

```python
#Initialisierung

x = 0.2     #Inputwert
y = x       #Erwarteter Ausgangswert y=x
eta = 1.0   #Lernrate

weight = -10
weights = []
errors = []
w_deltas = []
steps = []
```

```python
#Tabelle

print("Input x = {:.6f}, Gewünschter Output y = {:.2f}".format(x,y))
print("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}".format('Iter', 'x', 'w',
                                                        'n_input', 'act', 'y_hat', 'y', 'E', 'E_', 'w_delta'))

for step in range(10):
    net_i = weight*x                #Neuronen Input x*w
    activation = func_id(net_i)     #Aktivierungsfunktion y=x*w
    y_hat = activation              #Neuronen Output
    error = 0.5*(y-y_hat)**2        # Fehler zwischen Output und erwartetem Wert
    derivative = (-1.0)*x*(y-y_hat) #erste Ableitung - Gradient der Fehlerfunktion
    w_delta = (-1.0)*derivative*eta #Verbesserung für Wichtung w
    #weights.append(weight)
    #errors.append(error)
    #w_deltas.append(w_delta)
    #steps.append(step)

    if step % 1 == 0:
        print("{}\t{}\t{:6.2f}\t{:5.2f}\t{:5.2f}\t{:5.2f}\t{:.2f}\t{:.4f}\t{:.4f}\t{:.4f}"
              .format(step, x, weight, net_i, activation, y_hat, y, error, derivative, w_delta))

    weight += w_delta
```

```python
#Tabelle
weight = -10
print("Input x = {:.6f}, Gewünschter Output y = {:.2f}".format(x,y))
print("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}".format('Iter', 'x', 'w',
                                                        'n_input', 'act', 'y_hat', 'y', 'E', 'E_', 'w_delta'))

for step in range(121):
    net_i = weight*x
    activation = func_id(net_i)
    y_hat = activation
    error = 0.5*(y-y_hat)**2
    derivative = (-1.0)*x*(y-y_hat)
    w_delta = (-1.0)*derivative*eta
    weights.append(weight)
    errors.append(error)
    w_deltas.append(w_delta)
    steps.append(step)

    if step % 10 == 0:
        print("{}\t{}\t{:6.2f}\t{:5.2f}\t{:5.2f}\t{:5.2f}\t{:.2f}\t{:.4f}\t{:.4f}\t{:.4f}"
              .format(step, x, weight, net_i, activation, y_hat, y, error, derivative, w_delta))

    weight += w_delta
```

```python
#Plot

fig, ax1 = plt.subplots()
ax1.plot(weights, errors, label = "Fehler")
ax1.plot(weights, w_deltas, label = "w_deltas")
ax1.set_title('Gradientenabstieg')
legend = ax1.legend(loc = 'best', fancybox = True, framealpha=0.8)
plt.grid()
plt.xlabel('Gewicht w')
plt.show()
```

```python
#Plot

fig, ax1 = plt.subplots()
ax1.plot(steps, errors, label = "Fehler")
ax1.plot(steps, w_deltas, label = "w_deltas")
ax1.set_title('Gradientenabstieg')
legend = ax1.legend(loc = 'best', fancybox = True, framealpha=0.8)
plt.grid()
plt.xlabel('Iterationsschritte')
plt.show()
```
