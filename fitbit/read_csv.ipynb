```python
df3 = pd.read_csv('hourlySteps_merged.csv')
df3.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('How much is the fish', fontsize=14)
sns.countplot(x=df3['Id'], data=df3, palette=('YlOrRd'))
plt.xlabel("Id", fontsize=12)
plt.ylabel("Datenpunkte", fontsize=12)
plt.xticks(fontsize=12, rotation = 90)
plt.yticks(fontsize=12)
plt.show()
```

```python
category_counts = df3['Id'].value_counts()
grouped = df3.groupby('Id')
sum = grouped['StepTotal'].sum()
df3_id=df3['Id'].unique()
plt.figure(figsize=(10,8))
plt.title('Anzahl der Schritte pro Device', fontsize=14)
sns.barplot(x=sum, y=df3_id, orient = 'h')
plt.show()
```
